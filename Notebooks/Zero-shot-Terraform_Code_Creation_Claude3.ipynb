```python
import pandas as pd
import time
import re
import warnings
import json
warnings.filterwarnings('ignore')
import os
import scipy.spatial
import anthropic
#from anthropic import Anthropic, HUMAN_PROMPT, AI_PROMPT
warnings.filterwarnings("ignore")
```

```python
client = anthropic.Anthropic(
    # defaults to os.environ.get("ANTHROPIC_API_KEY")
    api_key="",
)
```

```python
df=pd.read_csv('/content/Combined_Terraform_File_validiac_validated_userquery_generated_sabya.csv')
del df['Unnamed: 0']
df
```

```python
prompt="""You are a Terraform consultant with expertise in AWS, Azure, and GCP. Your task is to develop a Terraform configuration based on the provided user query for the specified cloud provider.

- **User Query:** {}
- **Cloud Provider:** {}

This Terraform configuration must include modules for essential components and design the modules to be reusable and easy to maintain.

Adhere to the industry best practices for security and efficiency throughout the configuration. The output Terraform config should be in a structured format that is easy to understand and implement.
"""
```

```python
result=pd.DataFrame(columns=['Category','cp','Config','Userquery','Generated_Terraform_code_Claude'])
for index, row in df.head(2).iterrows():
  Category=row['Category']
  cp=row['cp']
  Config=row['Config']
  Userquery=row['Userquery']
  print(f"{Userquery}")
  input_query=prompt.format(Userquery,cp)
  #input_query="Give me terraform code for {} for the user input : {}".format(Provider,user_query)
  response_claude = anthropic.completions.create(
  model="claude-2",
  max_tokens_to_sample=300,
  prompt=f"{HUMAN_PROMPT} {input_query} {AI_PROMPT}",
)
  updated_response_claude = response_claude.completion
  new_row = {"Category":Category,'Provider':cp,"Config":Config,"Userquery":Userquery,"Generated_Terraform_code_Claude":updated_response_claude}
  new_df = pd.DataFrame([new_row])
  result = pd.concat([result, new_df], ignore_index=True)
result.head()
```

```python
result=pd.DataFrame(columns=['Category','cp','Config','Userquery','Generated_Terraform_code_Claude'])
for index, row in df.head(2).iterrows():
  Category=row['Category']
  cp=row['cp']
  Config=row['Config']
  Userquery=row['Userquery']
  print(f"{Userquery}")
  input_query=prompt.format(Userquery,cp)
  message = client.messages.create(
    model="claude-3-opus-20240229",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": input_query}
    ]
)
  updated_response_claude = message.content[0].text
  print(updated_response_claude)
  new_row = {"Category":Category,'cp':cp,"Config":Config,"Userquery":Userquery,"Generated_Terraform_code_Claude":updated_response_claude}
  new_df = pd.DataFrame([new_row])
  result = pd.concat([result, new_df], ignore_index=True)
result.head()
```
