```python
import pandas as pd
from openai import OpenAI
import time
import re
import warnings
import json
warnings.filterwarnings('ignore')
```

```python
MAX_RETRIES=2
SLEEP_DURATION=2
API_KEY=""""""
client = OpenAI(
    api_key=API_KEY,
)
```

```python
prompt=""""As an AI specialized in analyzing Terraform code : {} \n\n , your task is to examine the provided Terraform code and discern the complete set of user needs and intentions it addresses. Reverse engineer the Terraform code to derive only a single comprehensive non-technical user query summarised in one sentence, the sentence should start with "Please generate a Terraform configuration for ". The query should represent a full and distinct requirement or input that could have led to the creation of this specific Terraform configuration. The summary should be concise and understandable to a non-technical audience, adhering strictly to the information presented in the code.
"""
```

```python
def base_gpt4_model(query):
    for _ in range(MAX_RETRIES):
        try:
            response = client.chat.completions.create(
                #model='gpt-4-turbo',
                model='gpt-4o',
                messages=[
                    {"role": "user", "content": "{}".format(query)},
                ],
                temperature=0,
                max_tokens=200,
                top_p=0.95,
                frequency_penalty=0,
                presence_penalty=0,
                stop=None
                )
            return(response.choices[0].message.content)
        except Exception as e:
            if _ == MAX_RETRIES - 1: 
                time.sleep(SLEEP_DURATION) 
                return "An error occurred: {}".format(e)
```

```python
def base_gpt3_5_model(query):
    for _ in range(MAX_RETRIES):
        try:
            response = client.chat.completions.create(
                model='gpt-3.5-turbo-0125',
                messages=[
                    {"role": "user", "content": "{}".format(query)},
                ],
                temperature=0,
                max_tokens=200,
                top_p=0.95,
                frequency_penalty=0,
                presence_penalty=0,
                stop=None
            )
            return(response.choices[0].message.content)
        except Exception as e:
            if _ == MAX_RETRIES - 1: 
                time.sleep(SLEEP_DURATION) 
                return "An error occurred: {}".format(e)
```

```python
result_ben = pd.DataFrame(columns=['Category','Cloud_provider','user_query','Terraform_Code'])

for index, row in df_input.head(len(df_input)).iterrows():
    Category=row['Category']
    Cloud_provider=row['Cloud_provider']
    user_query=row['user_query']
    Terraform_Code=row['Terraform_Code']
    print(Category)
    try:
        updated_prompt=prompt.format(Terraform_Code)
        res=base_gpt4_model(updated_prompt)
        # res_updated = json.loads(res)
        # user_query_1=res_updated['userQuery1']
        # user_query_2=res_updated['userQuery2']
        # user_query_3=res_updated['userQuery3']
        result_ben = result_ben.append({
              'Category':Category,
              'Cloud_provider':Cloud_provider,
              'Terraform_Code':Terraform_Code,
              'user_query':res
              # 'user_query_1': user_query_1,
              # 'user_query_2':user_query_2,
              # 'user_query_3':user_query_3

          }, ignore_index=True)
        result_ben.to_excel('Research_Analysis_Base_data_filtered_new.xlsx')
    except Exception as error:
        print(error)
```
